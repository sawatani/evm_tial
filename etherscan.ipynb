```python
import requests

headers = {
    "User-Agent": "something" # これをつけないと BSC で失敗する
}
params = {
    'module': 'account',
    'action': action_erc721,
    'address': address_a
}
http_res = requests.get(etherscan_testnet['bsc'], headers=headers, params=params)
print(http_res.url)
print(http_res.status_code)
res = http_res.json()
if res['status'] != '1':
  raise Exception(res['message'])
print(res['message'])
```
